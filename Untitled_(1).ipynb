```python
import os
import glob
from PIL import Image
import cv2

# Paths
dataset_path = "/content/drowsiness_dataset/train"
output_path = "/content/YOLODataset"
image_id = 0
class_map = {'Closed_Eyes': 0, 'Open_Eyes': 1}

# Clear and recreate YOLO directories
!rm -rf /content/YOLODataset
os.makedirs(f"{output_path}/images/train", exist_ok=True)
os.makedirs(f"{output_path}/labels/train", exist_ok=True)

# Convert and label each image
for cls in class_map.keys():
    cls_path = os.path.join(dataset_path, cls)
    if not os.path.exists(cls_path):
        print(f"❌ Missing folder: {cls_path}")
        continue

    image_paths = glob.glob(cls_path + '/*.png')
    print(f"📂 {cls}: {len(image_paths)} images")

    for img_path in image_paths:
        try:
            filename = f"{image_id}.jpg"
            out_img_path = f"{output_path}/images/train/{filename}"

            # Convert PNG to RGB JPG
            im = Image.open(img_path).convert('RGB')
            im.save(out_img_path)

            # Image size for YOLO label
            img = cv2.imread(out_img_path)
            h, w, _ = img.shape

            # YOLO label format: class_id x_center y_center width height (normalized)
            label_line = f"{class_map[cls]} 0.5 0.5 1.0 1.0\n"
            label_path = f"{output_path}/labels/train/{image_id}.txt"
            with open(label_path, 'w') as f:
                f.write(label_line)

            image_id += 1
        except Exception as e:
            print(f"⚠️ Error with {img_path}: {e}")
```

```python
from IPython.display import clear_output

def handle_frame(data_url):
    # Step 1: Decode base64 webcam image to OpenCV frame
    header, encoded = data_url.split(",", 1)
    decoded = b64decode(encoded)
    np_arr = np.frombuffer(decoded, np.uint8)
    frame = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)

    # Step 2: Run YOLOv5 model inference
    results = model(frame)  # 🔥 This sends webcam frame to the model!

    # Step 3: Draw bounding boxes, labels on the frame
    results.render()  # Modifies results.ims[0] in-place

    # Step 4: Convert the result to an image and display
    img = Image.fromarray(results.ims[0])  # Convert numpy to PIL Image
    clear_output(wait=True)
    display(img)  # Show annotated image in output cell
```
